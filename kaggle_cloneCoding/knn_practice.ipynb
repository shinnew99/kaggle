KNN, K-nearest Neighborhood, K-최근접 이웃 알고리즘은 가장 간단한 머신러닝 알고리즘으로, 분류(Classification) 알고리즘이다. 
비슷한 특성을 가진 데이터는 비슷한 범주에 속하는 경향이 있다는 가정을 전제로 한다. <br>

이름처럼, 데이터가 분포해 있을 때 서로 인접한 거리내에 있는 점들을 묶어서 유사도를 보는 알고리즘이다. <br>
 <br>
주변의 가장 가까운 K개의 데이터를 보고 데이터가 속할 그룹을 판단하는 알고리즘이 K-NN 알고리즘이다. 따라서 거리 계산이 필수인데, 이때 사용하는 거리 계산방법은 유클리디안을 사용한다. <br>

KNN의 특징 중 하나는 K의 값에 따라 분류가 달라질 수 있다는 점이다. K는 홀수값으로 설정해준다. 만약 짝수로 줬을 때, 범위 원에 들어온 데이터 개수도 짝수가 된다. 최악의 경우는 (예를 들어 2 vs 2) 동률의 투표를 해야하는 상황이 되어버려 어느 레이블에 해당되는지 판단을 하기 어려워 진다. 따라서 일반적으로 K값을 홀수로 준다.

```python
# 빙어 14마리
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0] 
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
# from K_NN_data import * 
import matplotlib.pyplot as plt 

plt.scatter(bream_length, bream_weight) 
plt.scatter(smelt_length, smelt_weight) 
plt.xlabel('length') 
plt.ylabel('weight') 
plt.show()
```

```python
length = bream_length + smelt_length 
weight = bream_weight + smelt_weight
```

```python
fish_data = [[l, w] for l, w in zip(length, weight)]
```

```python
fish_target = [1] * 35 + [0] * 14
```

```python
from sklearn.neighbors import KNeighborsClassifier 
kn = KNeighborsClassifier()
```

```python
kn.fit(fish_data, fish_target) 
print(kn.score(fish_data, fish_target)) # 1.0
```

KNN의 특징은 훈련이 따로 필요 없다는 것. 훈련데이터를 기반으로 모델을 만들고 테스트 데이터로 테스트를 하는 방식이다. (<-- 사실상 모든 머신/딥러닝 알고리즘이 다 그렇다) 그냥 훈련 데이터를 저장하는 게 훈련의 전부이다. 새로운 데이터가 주어지면 그제야 주변의 k개 데이터를 보고 새로운 데이터를 분류해준다. 따라서 사전 모델링이 필요 없는 real-time 예측이 이루어진다. 

```python
kn = KNeighborsClassifier(n_neighbors=49)  # K값을 어떤걸로 바꿔도 무관
kn.fit(fish_data, fish_target) 
print(kn.score(fish_data, fish_target)) # 0.7142857142857143
```

### 3. 다른 데이터로 실습
위에선 안해줬기 때문에 훈련데이터와 테스트데이터를 분리해줘야한다.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# Assign colum names to the dataset
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']

# Read dataset to pandas dataframe
dataset = pd.read_csv(url, names=names)

dataset.head()
```

```python
# Preprocessing
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 4].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)

#Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

# Training and Predictions
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

# Evaluating the Algorithm
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
